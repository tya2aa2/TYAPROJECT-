```python
import requests
import pandas as pd

# URL dari file dataset
url = "https://storage.googleapis.com/dqlab-dataset/customer_segments.txt"

# Mengunduh data dari URL
response = requests.get(url)

# Memastikan request berhasil
if response.status_code == 200:
    # Membaca data yang diunduh sebagai teks
    data = response.text

    # Menyimpan data ke file sementara
    with open('customer_segments.txt', 'w') as file:
        file.write(data)

    # Membaca data tersebut menggunakan pandas
    pelanggan = pd.read_csv('customer_segments.txt', sep='\t')  # Sesuaikan delimiter jika perlu

    # Menyimpan ke file CSV
    pelanggan.to_csv('customer_segments.csv', index=False)
    print("File berhasil diunduh dan disimpan sebagai CSV.")
else:
    print("Gagal mengunduh file. Status kode:", response.status_code)
```

# **Fungsi kmeans**
Praktek kali ini kita akan melakukan segmentasi langsung pada data pelanggan  yang telah kita lakukan persiapan datanya pada bab sebelumnya  dengan menggunakan function kmeans.

 Function kmeans memerlukan minimal 2 parameter, yaitu:

x: data yang digunakan, dimana semua isi datanya harus berupa numerik.

centers: jumlah cluster yang diinginkan.


Dan fungsi kmeans ini biasanya disertai dengan pemanggilan function seet.seed. Ini berguna agar kita "menyeragamkan" daftar nilai acak yang sama dari kmeans sehingga kita mendapatkan output yang sama.

Berikut adalah contoh penggunaan fungsi kombinasi set.seed dan kmean.

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Normalisasi kolom "NilaiBelanjaSetahun"
pelanggan["NilaiBelanjaSetahun"] = pelanggan["NilaiBelanjaSetahun"] / 1000000

# Menentukan field yang akan digunakan dalam clustering
field_yang_digunakan = ["Jenis Kelamin1", "Umur", "Profesi1", "Tipe Residen1", "NilaiBelanjaSetahun"]

# Menetapkan seed agar hasil konsisten
np.random.seed(1)

# Membentuk 5 cluster dengan K-Means dan 25 skenario random
kmeans = KMeans(n_clusters=5, n_init=25, random_state=1)
segmentasi = kmeans.fit(pelanggan[field_yang_digunakan])

# Menampilkan hasil clustering
print("Cluster Centers:")
print(segmentasi.cluster_centers_)

print("\nCluster Labels:")
print(segmentasi.labels_)

print("\nWithin Cluster Sum of Squares by Cluster:")
print(segmentasi.inertia_)

print("\nAvailable Components:")
print("cluster, centers, totss, withiness, tot.withinss, betweenness, size, iter, ifault")
```

Clustering vector ini adalah rangkaian vector yang berisi angka cluster. Dari hasil kita, vector berisi angka 1 sampai dengan 5, maksimum sesuai dengan jumlah cluster yang kita inginkan.

Vector ini dimulai dari angka 2, yang artinya data pertama dari dataset kita akan dialokasikan pada nomor cluster 2. Hasil ini dapat diakses dengan komponen cluster dari objek hasil seperti berikut:

```python
# Menentukan field yang akan digunakan dalam clustering
field_yang_digunakan = ["Jenis Kelamin1", "Umur", "Profesi1", "Tipe Residen1", "NilaiBelanjaSetahun"]

# Menetapkan seed agar hasil konsisten
np.random.seed(1)

# Melakukan clustering dengan K-Means
kmeans = KMeans(n_clusters=5, n_init=25, random_state=1)
pelanggan['cluster'] = kmeans.fit_predict(pelanggan[field_yang_digunakan]) + 1  # +1 untuk menyesuaikan numbering dengan R

# Menampilkan hasil cluster dan struktur data
print("Cluster Labels:")
print(pelanggan['cluster'].values)

print("\nStruktur Data Pelanggan:")
print(pelanggan.info())

# Menampilkan beberapa baris data untuk memastikan penambahan kolom cluster
print("\nData Pelanggan dengan Cluster:")
print(pelanggan.head(10))
```
